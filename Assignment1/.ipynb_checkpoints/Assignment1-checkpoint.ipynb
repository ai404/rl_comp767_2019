### a) summarize the main results in the paper

The article performs an empirical analysis of 3 algorithms: Action Elimination Sampling (AE), Upper Confidence Bound (UCB) and LUCB (a variation of UCB) for a bandit problem of 6 arms. 

AE algorithm simply samples each arm *r* times. When it is confident enough that an arm is not the best arm it “throws” it out and stops sampling it. As it throws arms out, it gets closer to finding the optimal arm. The algorithm ends when there’s one arm left.

For UCB, each arm is initially sampled once. After that initial sampling, the algorithm samples the arms by selecting the arm whose estimated value and upper confidence bound is greatest. As an arm is sampled more and more, the upper confidence bound becomes tighter to the estimated value. Another way of looking at this is by seeing it as if the error of the estimation decrease as an arm is more and more sampled. Arms with a greater possible margins of error will be prioritized for sampling, id their estimated value are close to other arms.

Like UCB, in LUCB we sample each arm initially, than at each step we sample two arms, one that maximizes the upper bound of the estimates and another one that maximizes the estimates. The algorithm terminates when the difference between the estimates provided with the most sampled two arms is sufficiently high.

It turns out that LUCB has a more logical approach than UCB. In fact, UCB tends to sample the
best arm more than second best arm before being confident enough, contrary to LUCB which keeps sampling both arms until a certain confidence level is reached and then throws out the second best arm.

Also, the article established upper bounds for each algorithm on the maximum number of “pulls” the algorithm needs to achieve convergence. These bounds depends on the parameters of each algorithm. These bounds can be found on page 3 of the article in section *2 Bounding total number of measurements* of each algorithm . 

The article points out that the three algorithms have a very similar complexity level, yet it also shows how the algorithms differs. The AE algorithm tends to eliminate an arm at a time, hence it is slower than the other algorithms in identifying the best arm. On the other side, UCB and LUCB find the best arm very quickly with a huge gap between the probabilities of choosing the best arm and the the other arms. In terms of stopping time, LUCB showed to have the best results. 

At the end of the algorithms of AE and LUCB, the algorithm selects the best arm with probability of 1, while UCB selects the best arm with probability of about 0.9. This shows that UCB is not entirely certain and is still alternating with exploration and exploitation.

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from multiprocessing import Pool
import os
```

```python
# A class to simulate a Multi-armed bandit problem
class Environment:
    def __init__(self,means,variance = 0.1):
        self.means = means
        self.n_actions = len(means)
        self.sd = np.sqrt(variance)
    
    def get_actions(self):
        return np.arange(self.n_actions)

    def get_reward(self,i):
        return np.random.normal(self.means[i],self.sd)
    
    def plot(self):
        plt.figure(figsize=(10,7))
        data = np.random.normal(loc = self.means,scale=self.sd,size=(2000, self.n_actions))
        plt.xlabel("Arms")
        plt.ylabel("Value distribution")
        plt.violinplot(data,showmeans=True)
        
```

```python
# helper functions
def u(t,delta,eps):
    return (1+np.sqrt(eps)) * np.sqrt((1+eps) * np.log( np.log( (1+eps)*t + 2 ) /delta )/(2*t) )

def alpha(beta,n,delta):
    x = ((2+beta)/beta)**2
    return x*(1 + np.log(2*np.log(x*n/delta))/np.log(n/delta))

# calculate the probabilites for each timestep
def smoothing_probs(actions_history,n_actions,look_back=2):
    length = actions_history.shape[1]
    probabilities = np.zeros((length-look_back,n_actions))

    for i in range(look_back, length):
        for action in env.get_actions():
            probabilities[i-look_back,action] = np.mean(actions_history[:,i-look_back:i] == action)
    return probabilities
  
# plot the graphs of probabilities
def plot(probabilities,env,show_means=True):
    n_actions = env.n_actions
    plt.figure(figsize=(10,7))
    H1 = np.sum((env.means[0] - np.array(env.means[1:]))**-2)

    for action in range(n_actions):
        title = f"mu = {env.means[action]}" if show_means else f"arm {action+1}"
        plt.plot(np.arange(probabilities.shape[0])/H1,probabilities[:,action],label=title)

    plt.ylabel('P(It==i)')
    plt.xlabel('Number of pulls (units of H1)')
    plt.legend()

# convergence condition for ucb algorithm
def ucb_stop_condition(iteration,n_actions,T,beta,delta):
    if iteration < n_actions:
        return False

    for action in np.arange(n_actions):
        mask = np.ones_like(T,dtype=np.bool)
        mask[action] = False

        if T[action] > alpha(beta,n_actions,delta)* np.sum(T[mask]):
            return True
    return False

# functions to run experiments with multi processing
def experiment_wrapper(args):
    return args[0](args[1],**args[2])

def run_experiment(environment,algo,env_params,algo_params,n_runs = 100,n_cpu=None):
    if n_cpu is None:
        n_cpu = os.cpu_count()

    with Pool(n_cpu) as p:
        actions_history = list(tqdm(p.imap(experiment_wrapper, ((algo,environment(**env_params),algo_params) for _ in range(n_runs))), total=n_runs))
    return np.array(actions_history)
```

#### UCB Algorithm

```python
def ucb_algorithm(env,max_iters = 2500, eps = 0.01, delta = .1, beta = 1.66):

    n_actions = env.n_actions
    
    estimates = np.zeros(n_actions)
    actions = np.arange(n_actions)
    T = np.zeros(n_actions)

    run_actions_hist = []
    best_action = None

    for i in range(max_iters):
        if not ucb_stop_condition(i,n_actions,T,beta,delta):
            if i<n_actions:
                best_action = actions[i]
            else:
                values = estimates.copy()
                for action in range(n_actions):
                    values[action] = values[action] + (1+beta)*u(T[action],delta/n_actions,eps)
                best_action = np.argmax(values)
                run_actions_hist.append(best_action)

            r = env.get_reward(best_action)
            T[best_action]+=1
            estimates[best_action] = estimates[best_action] + (r - estimates[best_action]) / T[best_action]
        else:
            run_actions_hist.append(best_action)
    return run_actions_hist
```

#### LUCB Algorithm

```python

def lucb_algorithm(env,max_iters = 2500, eps = 0.01, delta = .1, beta = 1.66):

    n_actions = env.n_actions

    estimates = np.zeros(n_actions)
    T = np.zeros(n_actions)
    actions = np.arange(n_actions)
    np.random.shuffle(actions)

    run_actions_hist = []
    best_action = None
    i = 0
    while i < max_iters:
        if i<n_actions:
            action = actions[i]
            r = env.get_reward(action)
            T[action] += 1
            estimates[action] = estimates[action] + (r - estimates[action]) / T[action]
            run_actions_hist.append(action)

        elif best_action is None:
                action1 = np.argmax(estimates)
                r1 = env.get_reward(action1)
                T[action1]+=1
                estimates[action1] = estimates[action1] + (r1 - estimates[action1]) / T[action1]


                values = estimates.copy()
                for action in range(n_actions):
                    values[action] = values[action] + u(T[action],delta/n_actions,eps)

                values2 = values.copy()
                values2[action1] = np.min(values2)-1
                action2 = np.argmax(values2)


                r2 = env.get_reward(action2)

                ht_c = u(T[action1],delta/n_actions,eps)
                lt_c = u(T[action2],delta/n_actions,eps)

                if estimates[action1] - ht_c > estimates[action2] + lt_c:
                    best_action = action1
                    continue
                T[action2]+=1

                estimates[action2] = estimates[action2] + (r2 - estimates[action2]) / T[action2]

                run_actions_hist.append(action1)
                run_actions_hist.append(action2)
                i+=2
                continue
        else:
            run_actions_hist.append(best_action)
        i+=1
    return run_actions_hist
```

#### Action Elimination

```python
def ae_algorithm(env,r=1,max_iters = 2500, eps = 0.01, delta = .1, beta = 1.66):

    n_actions = env.n_actions

    estimates = np.zeros(n_actions)
    T = np.zeros(n_actions)
    valid_actions = np.arange(n_actions)
    np.random.shuffle(valid_actions)

    run_actions_hist = []
    i = 1
    while i <max_iters:
        if len(valid_actions) ==1:
            run_actions_hist.append(valid_actions[0])
            i+=1
            continue
        for _ in range(r):
            for a in valid_actions:
                if i >=max_iters:
                    break
                reward = env.get_reward(a)
                T[a]+=1
                estimates[a] = estimates[a] + (reward - estimates[a]) / T[a]
                run_actions_hist.append(a)
                i+=1

        f_c = lambda a: 2*u(i,delta/n_actions,eps)

        values = estimates.copy()
        action_ref = None
        for action in valid_actions:
            values[action] = values[action] + f_c(action)
            if action_ref is None or values[action]> values[action_ref]:
                action_ref = action
        valid_actions = [k for k in valid_actions if estimates[action_ref] - f_c(action_ref)<= estimates[k]+ f_c(k)]

    return run_actions_hist
```

### d) discuss in a short paragraph a concrete application in which you think regret optimization would be more useful than best arm identification


An interesting application we are considering is Ad optimization. The problem can be modeled, for a given keyword search or a query, as a multi-armed bandit where we try to select between a set of Ads (arms) the ones that we think will interest the user and show them amongst the search results. The action is to show an Ad or not within a search result and the reward here is whether the user clicked or not on the Ad we showed to him. Advertisers are targeting keywords, so we can only show the Ads that target the given keyword. Within search results we have limited spots for Ads and the Advertisers have also a limited budget to use. Let’s say we are trying to optimize the click through rate CTR, that is the number of times the Ad was clicked divided by the number of times it was shown within a search result. It is also the value of an action (show a specific Ad). The Advertiser is looking for a good campaign that generates clicks on its Ad. The Ad Company charges him for every display of the Ad. So here we have interest to show only the Ads that will generate clicks to serve both sides: avoid extra charges for the Advertiser and generate more profit for the Ad Company. 

Let’s suppose here that there’s so many more Advertisers than the number of available spots. If we use the Best Arm Identification, at the beginning all the Ads will be considered for the available spots. This may be aggressive on the budget of some advertisers since they will have a considerable number of displays but not enough clicks on their Ads to generate profit. Also, once we determine the Ads that generates the best CTRs, all the other Ads will be ignored for the given query. Consequently, when an Advertiser has edited his Ad and runs a new Ad campaign targeting the same keyword, we will not be able to reevaluate the new change if it has an effect on the CTR because our algorithm will never consider this Ad since it has already converged. Instead if we use Regret minimization we will be continuously considering all the Ads even those with low CTR. Those with high CTR will be more likely to show up on the search results, but also those with low CTR with a low probability. Doing so, we will be able to spot any change on weak Ads, monitor for any better Ad candidates or even track a new trend behavior among the people searching for that keyword without having to rerun the algorithm.

```python
def policyEvaluation(reward,values,p,discount,policy,maxDiff):
    #initialise diff to greater than maxDiff
    diff=maxDiff+1
    while diff>=maxDiff:
        diff=0
        #  sweep through all states
        for i in range(values.shape[0]):
            for j in range(values.shape[1]):
                temp=values[i,j]
                if (i==0 and j==0) or (i==0 and j==values.shape[1]-1):
                    values[i,j]=discount*values[i,j]
                else:
                    #create vector of possible future states
                    states=np.array([values[int(np.max((0,i-1))),j],values[i,int(np.min((j+1,values.shape[1]-1)))],values[int(np.min((values.shape[0]-1,i+1))),j],values[i,int(np.max((j-1,0)))]])
                    #create vector of rewards going to next state
                    rewards=np.array([reward[int(np.max((0,i-1))),j],reward[i,int(np.min((j+1,values.shape[1]-1)))],reward[int(np.min((values.shape[0]-1,i+1))),j],reward[i,int(np.max((j-1,0)))]])
                    #update state value
                    values[i,j]=p*(rewards[int(policy[i,j])]+discount*states[int(policy[i,j])])+(1-p)/4*(np.sum(rewards)+discount*np.sum(states))
                diff=np.max((np.abs(temp-values[i,j]),diff))  
    return values
                
def policyImprovement(policy,values,reward):
    new_policy=policy.copy()
    for i in range(values.shape[0]):
        for j in range(values.shape[1]):
            if (i==0 and j==0) or (i==0 and j==values.shape[1]-1):
                policy[i,j]=4
            else:
                #create vector of possible future states                                                                                               #create vector of possible future states
                states=np.array([values[int(np.max((0,i-1))),j],values[i,int(np.min((j+1,values.shape[1]-1)))],values[int(np.min((values.shape[0]-1,i+1))),j],values[i,int(np.max((j-1,0)))]])
                #create vectot of rewards going to next state
                rewards=np.array([reward[int(np.max((0,i-1))),j],reward[i,int(np.min((j+1,values.shape[1]-1)))],reward[int(np.min((values.shape[0]-1,i+1))),j],reward[i,int(np.max((j-1,0)))]])
                state_values=np.zeros(4) #create vector of future state values 
                #compute all possible futuristic state values
                for k in range(state_values.shape[0]):
                    state_values[k]=p*(rewards[k]+discount*states[k])+(1-p)/4*(np.sum(rewards)+discount*np.sum(states))
                #update value with max value and max diff value
                new_policy[i,j]=np.argmax(state_values)
    diff=bool(np.max(np.max(np.abs(policy-new_policy))))
    return new_policy, diff
```

```python
#initialise
discount=0.9
prob=[0.7,0.9]
N=[5,50]
maxDiff=0.001
dataPI_R=[]
dataPI_L=[]
for n in N:
    for p in prob:
        #initialise rewards
        rewards=np.zeros((n,n))
        rewards[0,0]=1
        rewards[0,n-1]=10
        #initialise policy 0=up, 1=right, 2=down, 3=left, 4 = terminal
        policy=np.random.randint(4,size=(n,n))
        policy[0,0]=4
        policy[0,n-1]=4
        #initialise values
        values=np.zeros((n,n))
        dataR=[]
        dataL=[]
        dataR.append(0)
        dataL.append(0)
        iter=0
        diff=True
        while diff:
            iter+=1
            values=policyEvaluation(rewards,values,p,discount,policy,maxDiff)
            policy,diff=policyImprovement(policy,values,rewards)
            L,R=calculateGreedyPol(values,rewards)
            dataR.append(R)
            dataL.append(L)

        #have a nice display for the policy
        outputPol=np.chararray((n,n), unicode=True)
        for i in range(len(actionsDisplay)):
            outputPol[policy==i]=actionsDisplay[i]
        print('n='+str(n)+', p='+str(p)+', iterations='+str(iter)+'\nFinal policy\n')
        print("\n".join([" ".join(line) for line in outputPol]))
        print('\n\n')

        dataPI_R.append(dataR)
        dataPI_L.append(dataL)
```

## Discussion

One can observe from the four graphs that all three methods converge to the same greedy policy value for the left and right bottoms squares. For all *n* and *p* the bottom left square’s greedy policy convergence value is constantly less than the right bottom square’s greedy policy convergence value. This can be explained by the fact that the bottom left square is further from the square giving a reward of 10, while the bottom right square is closer to that square. Since there is a probability of *p* that a chosen action leads to the desired state, it is uncertain that the agent will achieve the state with reward 10. Therefore the bottom right square, being closer than the bottom left square, will have a greater greedy policy value as is demonstrated by the graphs.

Another observation that can be made is that the rates of convergence is different for each method. The modified policy iteration converges faster than the policy iteration, and the policy iteration converges faster than the value iteration. 

To recap, the value iteration for one state evaluates all the possible state values using the Bellman’s equation and selects the greatest one. One update is when all states have been updated. The algorithm stops when there is a maximum change of less than epsilon (0.001 in our case) for all state values. It then outputs a greedy policy. 

Policy iteration is initiated with a given policy. That policy is then evaluated iteratively until the maximum change for all states is less than epsilon (0.001 in our case). This is called policy evaluation. The policy is improved at each state by calculating the greedy policy using the values from the policy evaluation. This step is called policy improvement. An update is composed of the policy evaluation and the policy improvement. When the policy does not change for any state, the algorithm stops and has converged.

Modified policy iteration consists of initiating a policy and doing value iteration *k* times (in our case we chose *k* to be 15). After the value iteration step is done, the policy is updated with the same method as the policy improvement step in policy iteration. This consist of one update. The algorithm ends when the updated policy is not changed from the previous update (just like in policy iteration). 

We can see that value iteration is dependant on epsilon. A smaller epsilon results in more updates since a smaller change in state values is required to stop the algorithm. Where as in policy iteration and modified policy iteration the algorithm stops when the policy no longer changes, it does not consider the change in value states as a stop criterion. This makes it converge faster since, as one can observe in table 1. Each algorithm was run 10 times since the initial policy is randomly determined and can have an effect on the number of updates required (especially for policy iteration). The average of these runs is shown in table 1. We also see that modified policy iteration does the least updates before converging. This is due to the fact that it does value iteration *k* times instead of policy evaluation as its first step. This in a way is already helping to optimize and select the greedy policy, as it chooses the best action to update the state values, where as in policy iteration it uses a policy that may not be optimal.

One can also observe that as *n* increases, the greedy policy value of the bottom square corners decrease. This is normal since they are further from the reward squares and thus less likely to reach them, because of the probability *p* of the action selected leading to the desired state.

Finally it is noticed that as *p* increases, so does the value of the bottom square corners. This is expected since the higher *p* is, the more probability that the agent will reach the desired state thus increasing the state values. When *p* is lower, there are more chances that a random action is taken, thus decreasing the greedy policy values of all the squares.
